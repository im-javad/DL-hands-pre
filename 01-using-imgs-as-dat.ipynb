# Converting a three-channel img to a single channel 

### The colors we get to 255, ther brighter the pixels become 

```python
imgData3 = np.array(plt.imread("../00-storage/image/sign.jpg"))
```

```python
gray_img = (imgData3[:, :, 0]//3) + (imgData3[:, :, 1]//3) + (imgData3[:, :, 2]//3)
```

```python
gray_img
```

```python
gray_img.shape
```

```python
plt.imshow(gray_img, cmap="gray")
```
