# Resizing dataset imgs

# Reshape dataset

we need to come up with a format that scikit-learn can accept

(1000, 5, 5) => (1000, 25(5*5)) 

These lines reshape the arrays from shape (n_samples, 5, 5) to (n_samples, 25(n_feature)) so that the data becomes compatible with scikit-learn. The value 25 flattens each 55 matrix into a 25-element vector, and -1 tells NumPy to automatically infer the number of samples.

```python
x_train = np.reshape(x_train_5by5, [-1,25])
x_test = np.reshape(x_test_5by5, [-1,25])
```
